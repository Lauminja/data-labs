```python
import pandas as pd
import numpy as np

import matplotlib as plt
```

```python
vehicles = pd.read_csv('../../lab-continuous-probability-distribution/your-code/vehicles.csv')
vehicles.head()
```

```python
vehicles.plot.scatter(x="City MPG", y="Highway MPG", alpha=0.2, grid=True)
```

```python
import seaborn as sns 
cmap = sns.cubehelix_palette(as_cmap=True) 
vehicles.plot.scatter(x="City MPG", y="Highway MPG", c="CO2 Emission Grams/Mile", cmap=cmap, alpha=0.2, grid=True)
```

```python
pd.plotting.scatter_matrix(vehicles, figsize=(20,10))
```

```python
m = pd.melt(vehicles[['City MPG', 'Highway MPG', 'CO2 Emission Grams/Mile']], id_vars='CO2 Emission Grams/Mile', var_name='MPG')
sns.lmplot(x='value',y='CO2 Emission Grams/Mile', hue='MPG', data=m,fit_reg=True)
```
