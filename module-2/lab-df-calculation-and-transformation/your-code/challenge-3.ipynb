## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
# How to aggregate the  amount_spent for unique customers?

cust_spend = orders.groupby(['CustomerID'])[['amount_spent']].sum()
cust_spend.sort_values('amount_spent')
```

```python
# How to select customers whose aggregated amount_spent is in a given quantile range?
print(cust_spend.quantile(0.95))
cust_spend[cust_spend.amount_spent > cust_spend.amount_spent.quantile(0.95)].sort_values('amount_spent')
```

```python
print(cust_spend.quantile(0.75))
cust_spend[(cust_spend.amount_spent > cust_spend.amount_spent.quantile(0.75)) 
           & (cust_spend.amount_spent < cust_spend.amount_spent.quantile(0.95))].sort_values('amount_spent')
```

```python
# How to label selected customers as "VIP" or "Preferred"?Â¶
cust_spend['Label'] = np.where(
                            (cust_spend.amount_spent > cust_spend.amount_spent.quantile(0.75))
                            & (cust_spend.amount_spent < cust_spend.amount_spent.quantile(0.95)), 'Preferred',
                            np.where(cust_spend.amount_spent > cust_spend.amount_spent.quantile(0.95), 'VIP',
                            'Non Pref'))

cust_spend.sort_values('amount_spent')
```

```python
labels = cust_spend.filter(['CustomerID', 'Label'], axis=1)
labels
orders = pd.merge(orders, labels, on='CustomerID', how='left')
orders
```
