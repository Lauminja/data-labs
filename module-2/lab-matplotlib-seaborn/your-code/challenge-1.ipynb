#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)

plt.subplot(1,2,1)
plt.plot(x,y,color='red')

plt.subplot(1,2,2)
plt.plot(x,z,c='green')

plt.show()
```

```python
from matplotlib import lines
lines.lineStyles.keys()
```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)
fig, [ay, az] = plt.subplots(1,2)

ay.plot(x,y,color='red', linewidth = 3, linestyle = '--')
ay.set_title('x&y')

az.plot(x,z,c='green', linewidth = 3, linestyle = ':')
az.set_title('x&z')

plt.show()
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
# your code here
fig, [ay, az] = plt.subplots(1,2, figsize=(12,12))

ay.plot(x,y,c='red', lw = 3, ls = '--')
ay.set_title('x&y')

az.plot(x,z,c='green', lw = 3, ls = ':')
az.set_title('x&z')

plt.show()
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# your code here
fig, [ay, az] = plt.subplots(1,2, figsize=(12,12))

ay.plot(x,y,c='r', lw = 3, ls = '--')
ay.set_title('x&y')
ay.set_xlabel('x label')
ay.set_ylabel('y label')

az.plot(x,z,c='g', lw = 3, ls = ':')
az.set_title('x&z')
az.set_xlabel('x label')
az.set_ylabel('y label')

plt.show()
```
