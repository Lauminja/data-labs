```python
import requests
from bs4 import BeautifulSoup

url = 'https://en.wikipedia.org/wiki/List_of_lists_of_lists'
html = requests.get(url).content
response = BeautifulSoup(html, "lxml")
response
```

```python
rows = response.find_all('div', {'class':'mw-parser-output'})
links = [i('a') for i in rows]
links
names = [link[0] for link in links]
names
#lists = [name.get('href').strip('/') for name in names]
#lists
```
