####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
#your code
html = requests.get(url).content
code = BeautifulSoup(html, "lxml")
````

````python
#Look up all classes relative to table data fields to include in the dataframe:
col = code.find_all('td',{'class':['tabev6', 'tabev1', 'tabev2', 'tabev3', 'tb_region', 'comment']})

#put all elements into a list, remove spaces and unnecessary characters
rows = [row.text.replace(u'\xa0', u' ').strip('earthquake').strip(' ') for row in col]

#split list into a list of lists, each sublist containing all data of one row
rows = [rows[9*i:9*(i+1)] for i in range(int(len(rows)/9) + 1)]

#join the latitute and longitude and their cardinal direction (N,E,S,W)

for items in rows:
    items[1:3] = [' '.join(items[1:3])]
    items[2:4] = [' '.join(items[2:4])]
rows
````

````python
import pandas as pd

colnames = [element.text for element in code.find_all('th', {'class':'th2'})]
colnames = colnames[1:]
colnames

data = rows[:20]

df = pd.DataFrame(data, columns=colnames)
df
````

#### Top 10 languages by number of native speakers stored in a Pandas Dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
#your code
html = requests.get(url).content
repos = BeautifulSoup(html, 'lxml')

table = repos.find_all('table',{'class':'wikitable sortable'})[1]

colnames = table.find_all('th')
colnames = [col.text.strip() for col in colnames]
colnames = colnames[:4]
colnames

rows = table.find_all('tr')
rows = rows[1:11]
rows = [row.text.strip().split("\n") for row in rows]
rows 

data = rows[0:9]
df = pd.DataFrame(data, columns=colnames)
df
````
