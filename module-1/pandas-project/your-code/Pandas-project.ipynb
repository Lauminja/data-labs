```python
#Extreme Values and Outliers:
stats = shark.describe().transpose()
stats['IQR'] = stats['75%'] - stats['25%']
stats
```

```python
outliers = pd.DataFrame(columns=shark.columns)

for col in stats.index:
    iqr = stats.at[col,'IQR']
    cutoff = iqr * 1.5
    lower = stats.at[col,'25%'] - cutoff
    upper = stats.at[col,'75%'] + cutoff
    results = shark[(shark[col] < lower) | 
                   (shark[col] > upper)].copy()
    results['Outlier'] = col #this added column will show the initial data set column name that contains the outlier
    outliers = outliers.append(results)
    
outliers.head()
```
