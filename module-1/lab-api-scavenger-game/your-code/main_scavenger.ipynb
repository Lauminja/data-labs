````python
url3 = 'https://api.github.com/repos/ironhack-datalabs/scavenger/contents/'

response3 = requests.get(url3, headers=header)
results3=json_normalize(response3.json())
results3.head()
````

````python
folder_list = results3['name']

content_list = []
for folder in folder_list[1:]: ### disregard first element in list to exclude .gitignore
    url_format = 'https://api.github.com/repos/ironhack-datalabs/scavenger/contents/{folder}'.format(folder=folder)
    response_folder = requests.get(url_format, headers=header)
    results4 = response_folder.json()
    content_list.extend(results4) 

content_list
````

````python
content2 = content_list.copy()  #make a copy in order to stop calling github and avoid being blocked due to too many requests

#To check that the list contains dictionaries:
# for x in content2:
#     print(type(x))
````
