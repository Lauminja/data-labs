````python
# Challenge 3: Hidden Cold Joke

#Using Python, call Github API to find out the cold joke contained in the 24 secret files in the following repo:
#https://github.com/ironhack-datalabs/scavenger

#The filenames of the secret files contain `.scavengerhunt` and they are scattered in different directories of this repo. 
#The secret files are named from `.0001.scavengerhunt` to `.0024.scavengerhunt`. They are scattered randomly throughout this repo. 
#You need to **search for these files by calling the Github API**, not searching the local files on your computer.
````

````python
url3 = 'https://api.github.com/repos/ironhack-datalabs/scavenger/contents/'

response3 = requests.get(url3, headers=header)
results3=response3.json()
results3=json_normalize(results3)
results3.head(10)
````

````python
folder_list = results3['name']

df=pd.DataFrame()
for folder in folder_list[1:]: ### disregard first element in list to exclude .gitignore
    url_format = 'https://api.github.com/repos/ironhack-datalabs/scavenger/contents/{folder}'.format(folder=folder)
    response_folder = requests.get(url_format, headers=header)
    results4 = response_folder.json()
    df = df.append(results4)

df
````

````python
#After you find out the secrete files:
#Sort the filenames ascendingly.

df2 = df[df['path'].str.contains('.scavengerhunt')].sort_values('name')
df2
````

````python
scavenger = df2['path']
scavenger
````

````python
#Get all the responses in one list:

responses = []
for doc in scavenger:
    content_link = 'https://api.github.com/repos/ironhack-datalabs/scavenger/contents/{}'.format(doc)
    response = requests.get(content_link, headers=header)
    print(response.content)
    responses.append(response)
````

````python
#Need to access each "download_url" key to be able to print out the string in file:
import json

downloads = [json.loads(x.content)["download_url"] for x in responses]
downloads
````

````python
# Get content of each file:
import requests
files = [requests.get(download).text.strip() for download in downloads]
````

````python
#Concatenate the strings in the array separating each two with a whitespace.
#Print out the joke.

quote = ' '.join(files)
quote
````
